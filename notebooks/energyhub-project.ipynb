For informations_households:
- `LCLid` household id
- `stdorToU` the kind of tarrif applied (ToU is the dynamic tarrif in days, Std is the classic fixed tarrif)
- `Acorn` the [ACORN](https://acorn.caci.co.uk/) group associated with the house category
- `Acorn_grouped` global classification of the ACORN (i.e. a fusion of difference ACORN groups)
- `file` name of the file in the different zip files to find the data of the household

#### A note on `stdorToU`
The data are two groups of customers. The first is a sub-group, of approximately 1100 customers, who were subjected to Dynamic Time of Use (dToU) energy prices throughout the 2013 calendar year period. The tariff prices were given a day ahead via the Smart Meter IHD (In Home Display) or text message to mobile phone. Customers were issued High (67.20p/kWh), Low (3.99p/kWh) or normal (11.76p/kWh) price signals and the times of day these applied. The dates/times and the price signal schedule is availaible as part of this dataset. All non-Time of Use customers were on a flat rate tariff of 14.228pence/kWh. The signals given were designed to be representative of the types of signal that may be used in the future to manage both high renewable generation (supply following) operation and also test the potential to use high price signals to reduce stress on local distribution grids during periods of stress.

The remaining sample of approximately 4500 customers energy consumption readings were not subject to the dToU tariff

```python
# add weather data
daily_dataset_2013 = pd.merge(daily_dataset_2013, daily_weather, how='left', left_on='day', right_on='time')
daily_dataset_2013.shape
```

```python
# get house and blocks for dates
Std_blocks = list(set(household_info[household_info['stdorToU'] == 'Std']["file"].values))
#ToU_blocks = list(set(household_info[household_info['stdorToU'] == 'ToU']["file"].values))
Std_houses = daily_dataset_2013[daily_dataset_2013['stdorToU'] == 'Std'].groupby('LCLid').count().reset_index()['LCLid'].values
#ToU_houses = daily_dataset_2013[daily_dataset_2013['stdorToU'] == 'ToU'].groupby('LCLid').count().reset_index()['LCLid'].values
```

```python
# number of houses drawing energy
len(Std_houses)
```

```python
# get half-hourly data for each type of tarrif in 2013

## Std Tarrifs
Std_block_list = []

# set datatype for energy column
dtype = {"energy(kWh/hh)":float}

for block in tqdm(Std_blocks):
    
    # read in current block
    curr_block = pd.read_csv(INPUT_PATH + DATA_PATH + f"halfhourly_dataset/halfhourly_dataset/{block}.csv",
                             dtype=dtype,
                             parse_dates=[1],
                             na_values="Null")
   

    # apply filters
    curr_block = curr_block[curr_block['tstp'].dt.year == 2013]
    curr_block = curr_block[curr_block["LCLid"].isin(Std_houses)]
    curr_block = curr_block[~curr_block['energy(kWh/hh)'].isna()]
    
    # convert string type to float
    curr_block['energy(kWh/hh)'] = curr_block['energy(kWh/hh)'].astype(float)
    
    
    # append current block to list blocks
    
    Std_block_list.append(curr_block)
    
# concat the list along the row axis    
Std_2013_hh = pd.concat(Std_block_list, axis=0)
Std_2013_hh = Std_2013_hh.rename(columns={'tstp':'time'})
Std_2013_hh.shape
```

```python
# grab UV data
UV_DATA_PATH = "UK-radiation/2013/"
uv_data_2013 = pd.read_csv(INPUT_PATH + UV_DATA_PATH + "midas-open_uk-radiation-obs_dv-202207_greater-london_00726_kenley-airfield_qcv-1_2013.csv",
                      sep=",",
                      header=0,
                      parse_dates=[0])

# remove last end row
uv_data_2013 = uv_data_2013.iloc[:-1, :]
uv_data_2013['time'] = pd.to_datetime(uv_data_2013['ob_end_time'])
hourly_irradiance_2013 = uv_data_2013[['time', 'glbl_irad_amt']]
hourly_irradiance_2013 = hourly_irradiance_2013.rename(columns={'glbl_irad_amt':'global irradiance'}) #KJ/m2
```

```python
hourly_irradiance_2013[hourly_irradiance_2013['time'].dt.month == 8].set_index('time').plot(style="-", figsize=(30,7))
```

```python
# merge dataframes on time
energy_df = pd.merge(hourly_demand_2013, hourly_weather_2013, on='time')
energy_df = pd.merge(energy_df, hourly_irradiance_2013, on='time')
energy_df = energy_df.drop(['apparentTemperature'], axis=1)

# check the values of the categorical cols
energy_df[['precipType','icon']].value_counts()
```

```python
energy_df.set_index('time').select_dtypes(include='number').plot(figsize=(30,12))
```

```python
# encode the categorical features and add to dataframe
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
categorical_cols = ohe.fit_transform(energy_df[['icon', 'precipType']])

# column names for the dataframe
categorical_col_names = [label for category in ohe.categories_ for label in category]
categoricals_df = pd.DataFrame(categorical_cols.toarray(), columns = categorical_col_names)
processed_df = pd.concat([energy_df, categoricals_df], axis=1)
processed_df = processed_df.set_index('time').select_dtypes(include='number')
processed_df.head()
```

```python
# generate heatmap for pearson correlation
pear_corr_matrix = processed_df.select_dtypes(include='number').corr()
plt.figure(figsize=(15,12))
sns.heatmap(pear_corr_matrix, annot=True, cmap='coolwarm')
plt.title('Pearson Correlation Heatmap')
plt.show()
```
